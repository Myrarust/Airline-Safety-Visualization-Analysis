```python
# Create a subset of the data that I am going to use for the heatmap
df3 = df2.drop(['Name  ', 'G', 'MIN', 'PTS', 'FGM', 'FGA', 'FTM', 'FTA', '3PM', '3PA', 'ORB', 'DRB', 'TRB',
                 'AST', 'STL', 'BLK', 'TO', 'PF'], axis=1)
```

```python
# Each of these values are on the same scale and is a percentage, so I want to multiple the columns by 100. 
df3["FGP"] = 100 * df3["FGP"]
df3["FTP"] = 100 * df3["FTP"]
df3["3PP"] = 100 * df3["3PP"]
```

```python
# Set figure size
from matplotlib.pyplot import figure
figure(figsize=(10, 15), dpi=80)

# Add a title for the viz
plt.title("NBA Per Game Performance", size=18)

# Define a color palette
cmap = sns.color_palette("Blues", as_cmap=True)

# create seabvorn heatmap with required labels
sns.heatmap(df3, xticklabels=x_axis_labels, yticklabels=y_axis_labels, annot=True, annot_kws={"size": 8}, cmap=cmap, fmt='g')
```

***Contour Chart***

```python
# Use TRB values for xlist and STL values for ylist
xlist = list(df2.TRB)
ylist = list(df2.STL)
```

```python
# Set up the data for the contour plot
X, Y = np.meshgrid(df2.TRB, df2.STL)
Z = X**2 + Y**2
```

```python
# Plot the contour chart
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
plt.show()
```

```python
# Import libraries
import descartes
import geopandas as gpd
from shapely.geometry import Point, Polygon
%matplotlib inline
```

```python
# Take a look at dataset 1
df = pd.read_csv("ex5-costcos-geocoded.csv")
df.head()
```

```python
# Use list comprehension to create points from the Lat/Long data
geometry = [Point(xy) for xy in zip(df.Longitude, df.Latitude)]

# Check to make sure it worked
geometry[:3]
```

```python
# Define the coordinate system
crs = {'init': 'epsg:4326'}

# Create our geometry dataframe
geo_df = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
geo_df.head()
```

```python
# Read in shape file of the US (retrieved from ArcGIS)
map = gpd.read_file("States_shapefile.shp")
```

```python
# Display the shape file
fig, ax = plt.subplots(figsize = (15, 15))
map.plot(ax=ax)
```

```python
# Okay, now plot the points ontop of the shapefile
fig, ax = plt.subplots(figsize = (15, 15))
map.plot(ax=ax, alpha=0.4, color="grey")

# Add a title
plt.title("Costco Locations in the US", size=18)

# Add points layer
geo_df.plot(ax=ax, markersize=3, color="red", marker="o")
```
